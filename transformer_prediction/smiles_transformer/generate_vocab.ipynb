```python
import pandas as pd
df = pd.read_table("chembl_24_1_chemreps.txt")
L = len(df)
print(L)
df.head()
```

```python
smiles = df["canonical_smiles"].values
to_drop = []
for i, sm in enumerate(smiles):
    if len(sm) > 100:
        to_drop.append(i)
    if df['chembl_id'][i] == "CHEMBL1201364":
        to_drop.append(i)
print(len(to_drop))
```

```python
df_drop = df.drop(to_drop)
df_drop.to_csv("data/chembl_24.csv", index=False)
```

```python
df_sol = pd.read_csv('data/datasets/freesolv/freesolv.csv')
df_sol.head()
```

```python
smiles = df_sol['smiles'].values
pro_sms = []
for sm in smiles:
    sm = ' '.join(list(sm))
    before = ['C l -', 'C l', 'O -', 'N +', 'n +', 'B r -', 'B r', 'N a +', 'N a', 'I -', 'S i']
    after = ['Cl-', 'Cl', 'O-', 'N+', 'n+', 'Br-', 'Br', 'Na+', 'Na', 'I-', 'Si']
    for b,a in zip(before, after):
        sm = sm.replace(b, a)
    pro_sms.append(sm)

df_sol['processed_smiles'] = pro_sms
df_sol.head()
```

```python
df_sol.to_csv('data/solubility.csv')
```
